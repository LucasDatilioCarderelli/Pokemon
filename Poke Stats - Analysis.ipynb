```python
import seaborn as sns

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (9,5)

# pokemons['HP'].value_counts()
# pokeHPmult10 = round(pokemons['HP'], -1).value_counts()

# pokemons['HP'].hist()
bins = [10,30,50,70,90,110,130,150,170,190,210,230,250]

plt.hist(pokemons['HP'], bins=bins)
# sns.distplot(pokemons.HP, bins=bins)
plt.title('DistribuiÃ§Ã£o por Hp')
plt.xlabel('Hp')
plt.ylabel('Un')

plt.xticks(bins)

plt.show()
```
